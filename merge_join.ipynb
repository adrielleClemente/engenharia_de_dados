Crie um dataframe único com a informação do GDP e quantidade de matrículas por nível de educação por ano.

O aumento das matrículas na educação primária, secundária e terciária acompanhou o aumento do GDP per capita nos últimos 15 anos na India?

## Transformando os Dataframes

```python
df_educ = df_educ[df_educ['Country']=='India']
series = ['Students enrolled in primary education (thousands)',
          'Students enrolled in secondary education (thousands)',
          'Students enrolled in tertiary education (thousands)']
```

```python
gpd = df_educ['Series'].isin(series) # True | False
df_educ = df_educ[gpd]
df_educ
```

```python
df_educ = df_educ.drop(['Region/Country/Area','Country','Source','Footnotes'],
                       axis=1)
df_educ
```

```python
# sem a retirada da vírgula, algumas células estavam dado erro.
df_educ['Value'] = df_educ['Value'].astype(str).str.replace(',','').astype(int)
```

```python
df_educ.info()
```

```python
df_educ['Series'] = df_educ['Series'].map({'Students enrolled in secondary education (thousands)':'primary',
                                           'Students enrolled in secondary education (thousands)':'secondary',
                                           'Students enrolled in tertiary education (thousands)':'tertiary'})
df_educ
```

```python
# por algum motivo na hora de renomear pra primary o df colocava NaN, então eu apenas substituí.
df_educ['Series'] = df_educ['Series'].fillna('primary')
```

```python
df_educ = pd.DataFrame(df_educ.groupby(['Year','Series'])['Value'].sum()).reset_index()
df_educ
```

```python
df_educ = df_educ.pivot(index = 'Year', columns = 'Series', values = 'Value') # para saber como cada Serie se comporta separadamente.
df_educ
```

```python
# sem os escapes a linha não era carregada no df.
gpd_per = df_gdp['Indicator Name'].str.contains('GDP per capita \\(current US\\$\\)')
df_gdp = df_gdp[gpd_per]
df_gdp
```

```python
# df_gdp.info()
```

```python
colunas = range(2003,2018)
df_gdp = pd.melt(df_gdp, id_vars=['Indicator Name'],
                value_vars = list(map(str, colunas)),
                var_name = 'Year', value_name = 'Value')
df_gdp
```

```python
df_gdp = df_gdp[['Year','Value']].groupby(['Year']).sum()
df_gdp['Value'] = df_gdp['Value'].round().astype(int)
df_gdp
```

```python
df_educ
```

```python
df_gdp.index = df_gdp.index.astype(int) # o index precisava ser marcado com int
df = df_gdp.merge(df_educ, on = 'Year',
                  how = 'outer',
                  suffixes = ('_gpd','_educ'))
df
```

## Gráfico com os Valores Reais

```python
plt.figure(figsize = (15,10))

x = df.index

y1 = df['Value']
y2 = df['primary']
y3 = df['secondary']
y4 = df['tertiary']

plt.plot(x, y1, label = 'GPD_Value')
plt.plot(x, y2, label = 'primary')
plt.plot(x, y3, label = 'secondary')
plt.plot(x, y4, label = 'tertiary')

plt.xticks(x)
plt.grid()
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('GPD x Matrículas')

plt.legend()
plt.show()
```

## Melhorando a Visualização do Gráfico

```python
df['Value'] = df['Value']*100
```

```python
plt.figure(figsize = (15,10))

x = df.index

y1 = df['Value']
y2 = df['primary'].interpolate(method = 'linear')
y3 = df['secondary'].interpolate(method = 'linear')
y4 = df['tertiary'].interpolate(method = 'linear')

plt.plot(x, y1, label = 'GPD_Value')
plt.plot(x, y2, label = 'primary', marker = 'o')
plt.plot(x, y3, label = 'secondary', marker = 'o')
plt.plot(x, y4, label = 'tertiary', marker = 'o')

plt.xticks(x)
# plt.grid()
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Produto Interno Bruto per capita x Matrículas')

plt.legend()
plt.show()
```
